```python
# 47. Check whether a number is strong number (sum of factorial of digits).
n = int(input("Enter the number: "))
def fact(num):
    prod = 1
    for i in range(1,num+1):
        prod*=i
    return prod
t = n
summ = 0
while t:
    summ = summ + fact(t%10)
    t = t//10
print("Strong Number" if summ==n else "Not a strong number")
```

```python
# 89. Simulate ATM: balance, withdraw, deposit, exit.


balance = 1000.0  
pin = "1234"      
exit_atm = False

print("Welcome to the ABC Bank ATM")

user_pin = input("Enter your PIN: ")
if user_pin != pin:
    print("Invalid PIN. Exiting.")
    exit_atm = True


while not exit_atm:
    print("\n-------------------------")
    print("1. Check Balance")
    print("2. Deposit Money")
    print("3. Withdraw Money")
    print("4. Exit")
    print("-------------------------")
    
    choice = input("Enter your choice (1-4): ")

    if choice == '1':
        
        print(f"\nYour current balance is: ${balance:.2f}")

    elif choice == '2':
        
        amount = float(input("Enter amount to deposit: $"))
        if amount > 0:
            balance += amount
            print(f"Successfully deposited ${amount:.2f}")
            print(f"New balance: ${balance:.2f}")
        else:
            print("Invalid deposit amount.")

    elif choice == '3':
       
        amount = float(input("Enter amount to withdraw: $"))
        if amount > 0:
            if amount <= balance:
                balance -= amount
                print(f"Successfully withdrew ${amount:.2f}")
                print(f"Remaining balance: ${balance:.2f}")
            else:
                print("Error: Insufficient funds.")
        else:
            print("Invalid withdrawal amount.")

    elif choice == '4':
        
        print("Thank you for using our ATM. Goodbye!")
        exit_atm = True

    else:
        print("Invalid choice. Please select 1-4.")
```

```python
# 90. Guess-the-number game (loop until correct).

secret_number = 1234
attempts = 0
guess = None
print("Welcome to the Guess the Number Game!")
print("I have picked a number between 1 and 100.")
while guess != secret_number:
    try:
        guess = int(input("Enter your guess: "))
        attempts += 1
        if guess < secret_number:
            print("Too low! Try again.")
        elif guess > secret_number:
            print("Too high! Try again.")
    except ValueError:
        print("Invalid input. Please enter an integer.")
print(f"Congratulations! You guessed the correct number ({secret_number}) in {attempts} attempts.")
```

```python
# 91. Find the first N numbers divisible by both 4 and 6.
n = int(input("Enter the n numbers: "))
count = 0
i = 1
while count<n:
    if i%4==0 and i%6==0:
        print(i)
        count+=1
    i+=1
```

```python
# 94. Find the smallest number ≥ N that is divisible by 12.

N = int(input("Enter a number: "))

if N % 12 == 0:
    result = N
else:
    result = N + (12 - N % 12)

print("The smallest number ≥ N divisible by 12 is:", result)
```

```python
# 95. Find the largest number ≤ N that is divisible by 9.
N = int(input("Enter a number: "))
if N % 9 == 0:
    result = N
else:
    result = N - (N % 9)
print("The largest number ≤ N divisible by 9 is:", result)
```

```python
# 96. Find the count of digits that are even in a number.
N = int(input("Enter a number: "))
count = 0
for digit in str(abs(N)):
    if int(digit) % 2 == 0:
        count += 1
print("Count of even digits:", count)
```

```python
# 97. Find the sum of digits at odd positions (from right).
N = input("Enter a number: ")
N = N[::-1]
sum_odd_pos = 0
for i in range(0, len(N), 2):
    sum_odd_pos += int(N[i])
print("Sum of digits at odd positions:", sum_odd_pos)
```

```python
# 98. Convert decimal to binary using loops.
N = int(input("Enter a decimal number: "))
if N == 0:
    print("Binary:", 0)
else:
    binary = ""
    num = N
    while num > 0:
        binary = str(num % 2) + binary
        num = num // 2
    print("Binary:", binary)
```

```python
# 99. Convert decimal to octal using loops.
N = int(input("Enter a decimal number: "))
if N == 0:
    print("Octal:", 0)
else:
    octal = ""
    num = N
    while num > 0:
        octal = str(num % 8) + octal
        num = num // 8
    print("Octal:", octal)
```

```python
# 100. Convert decimal to hexadecimal using loops (without hex())
decimal_num = int(input("Enter the decimal number: "))
if decimal_num == 0:
    print("0")
hex_chars = "0123456789ABCDEF"
hex_result = ""
current_num = decimal_num
while current_num > 0:
    remainder = current_num % 16
    hex_digit = hex_chars[remainder]
    hex_result = hex_digit + hex_result
    current_num = current_num // 16
print(f"Decimal {decimal_num} is Hexadecimal: {hex_result}")
```
